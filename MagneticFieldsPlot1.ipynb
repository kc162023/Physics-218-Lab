```python
import numpy as np
import matplotlib.pyplot as plt


%matplotlib inline


#Create arrays for your data
Current = np.array([-.0045,.0042,-.0123,.0133,-.0171,.0198,.0279,.0340,-.045,.0409])#Amps
tan_theta = np.array([-.1763,.1763,-.3640,.3640,-.5774,.5774,-.8391,.8391,-1.1918,1.1918])

#Create an array for your y-axis uncertainties
err_current = np.array([.0001,.0001,.0001,.0001,.0001,.0001,.0001,.0001,.0001,.0001])#Amps
#Reassign variables
x = tan_theta
y = Current
dy = err_current

#size the plot
plt.figure(figsize=(15,10))

#create scatter plot
plt.scatter(x, y, color='blue', marker='o')

#create labels
plt.xlabel('tan(theta)')
plt.ylabel('Current(Amps)')
plt.title('Current vs Tan(theta)')

#fitting to a 2nd degree polynomial
c,b,a=np.polynomial.polynomial.polyfit(x,y,2,w=dy)

#Annotate with values of A, B, C from best fit polynomial
plt.annotate('Slope = {value:.{digits}E}'.format(value=a, digits=3),
             (0.05, 0.9), xycoords='axes fraction')

plt.annotate('B = {value:.{digits}E}'.format(value=b, digits=3),
             (0.05, 0.85), xycoords='axes fraction')
             
plt.annotate('C = {value:.{digits}E}'.format(value=c, digits=3),
             (0.05, 0.8), xycoords='axes fraction')
#Create fit line
xnew = np.linspace(x.min(), x.max(), 300)
fit = a*xnew**2 + b*xnew +c

plt.scatter(xnew, fit, color='red')
plt.show()

print ("C = ",c , "B = ",b, "A = ",a)
```
